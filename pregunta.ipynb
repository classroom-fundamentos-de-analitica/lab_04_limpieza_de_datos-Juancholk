#h1 Transform string to float or int

```python
df = pd.read_csv("solicitudes_credito.csv", sep=";")
df.head()
```

```python
df.dtypes
```

```python
df.monto_del_credito =df.monto_del_credito.str.strip('$')
df['monto_del_credito'] = df['monto_del_credito'].str.replace(',', '')  # Quitar las comas
# Quitar ".0" seguido de ceros
df['monto_del_credito'] = df['monto_del_credito'].str.replace(r'\.0+$', '', regex=True)
df['monto_del_credito'] = df['monto_del_credito'].astype(int)  # Convertir a enteros
#df.monto_del_credito = df.monto_del_credito.astype(int)
df.monto_del_credito
```

```python
df.head()
```

```python
df.estrato = df.estrato.astype('category')
df.comuna_ciudadano = df.comuna_ciudadano.astype(int)
display(df.estrato.cat.categories,
        df,
        df.dtypes)
```

```python
#eliminar la columna unnamed
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
def invertir_fecha(fecha):
    partes = fecha.split('/')
    if len(partes[2]) == 4:
        return f"{partes[2]}/{partes[1]}/{partes[0]}"
    return fecha
```

```python
df['fecha_de_beneficio'] = df['fecha_de_beneficio'].apply(invertir_fecha)

# Convertir todas las fechas al formato datetime
df['fecha_de_beneficio'] = pd.to_datetime(df['fecha_de_beneficio'])


display(df.head(),
df.dtypes)
```

```python
import datetime as dt
today = dt.datetime.today()
today
```

```python
#verificar si la fecha de beneficio es mayor a la fecha actual
df[df['fecha_de_beneficio'] > today]
```
